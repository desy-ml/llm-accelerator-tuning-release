```python
import matplotlib.pyplot as plt
import numpy as np

from src.environments.reward import combiners, transforms
```

```python
abs_transform = transforms.Abs()
```

```python
linear_transform = transforms.Linear(good=0, bad=2e-3)
clipped_linear_transform = transforms.ClippedLinear(good=0, bad=2e-3)
softplus_transform = transforms.SoftPlus(good=0, bad=2e-3)
negexp_transform = transforms.NegExp(good=0, bad=2e-3)
sigmoid_transform = transforms.Sigmoid(good=0, bad=2e-3)
```

```python
errors = np.linspace(-3.5e-3, 3.5e-3, 200).tolist()

plt.plot(
    errors, clipped_linear_transform(abs_transform(errors)), label="clipped linear"
)
plt.plot(errors, softplus_transform(abs_transform(errors)), label="softplus")
plt.plot(errors, negexp_transform(abs_transform(errors)), label="negexp")
plt.plot(errors, sigmoid_transform(abs_transform(errors)), label="sigmoid")
plt.legend()
plt.show()
```

```python
errors = np.linspace(-3.5e-3, 3.5e-3, 200).tolist()

plt.plot(errors, linear_transform(abs_transform(errors)), label="linear")
plt.plot(
    errors, clipped_linear_transform(abs_transform(errors)), label="clipped linear"
)
plt.plot(errors, softplus_transform(abs_transform(errors)), label="softplus")
plt.plot(errors, negexp_transform(abs_transform(errors)), label="negexp")
plt.plot(errors, sigmoid_transform(abs_transform(errors)), label="sigmoid")
plt.legend()
plt.show()
```

```python
errors = np.linspace(-3.5e-3, 3.5e-3, 200).tolist()

correct_sigmoid_transform = transforms.Sigmoid(good=0, bad=2e-3)
reversed_sigmoid_transform = transforms.Sigmoid(good=2e-3, bad=0)

plt.plot(
    errors, correct_sigmoid_transform(abs_transform(errors)), label="correct sigmoid"
)
plt.plot(
    errors, reversed_sigmoid_transform(abs_transform(errors)), label="reversed sigmoid"
)
plt.legend()
plt.show()
```

```python
mean_combiner = combiners.Mean()
multiply_combiner = combiners.Multiply()
geometric_mean_combiner = combiners.GeometricMean()
min_combiner = combiners.Min()
max_combiner = combiners.Max()
l1_combiner = combiners.LNorm(1)
l2_combiner = combiners.LNorm(2)
smoothmax_combiner = combiners.SmoothMax(-5)
```

```python
errors = np.linspace(-3.5e-3, 3.5e-3, 200).tolist()

sigmoid_errors = sigmoid_transform(abs_transform(errors))
softplus_errors = softplus_transform(abs_transform(errors))
clipped_linear_errors = clipped_linear_transform(abs_transform(errors))

final = [
    mean_combiner([sigmoid, softplus, clipped_linear])
    for sigmoid, softplus, clipped_linear in zip(
        sigmoid_errors, softplus_errors, clipped_linear_errors
    )
]
plt.plot(errors, final, label="mean")

final = [
    multiply_combiner([sigmoid, softplus, clipped_linear])
    for sigmoid, softplus, clipped_linear in zip(
        sigmoid_errors, softplus_errors, clipped_linear_errors
    )
]
plt.plot(errors, final, label="multiply")

final = [
    geometric_mean_combiner([sigmoid, softplus, clipped_linear])
    for sigmoid, softplus, clipped_linear in zip(
        sigmoid_errors, softplus_errors, clipped_linear_errors
    )
]
plt.plot(errors, final, label="geometric mean")

final = [
    min_combiner([sigmoid, softplus, clipped_linear])
    for sigmoid, softplus, clipped_linear in zip(
        sigmoid_errors, softplus_errors, clipped_linear_errors
    )
]
plt.plot(errors, final, label="min")

final = [
    max_combiner([sigmoid, softplus, clipped_linear])
    for sigmoid, softplus, clipped_linear in zip(
        sigmoid_errors, softplus_errors, clipped_linear_errors
    )
]
plt.plot(errors, final, label="max")

final = [
    l1_combiner([sigmoid, softplus, clipped_linear])
    for sigmoid, softplus, clipped_linear in zip(
        sigmoid_errors, softplus_errors, clipped_linear_errors
    )
]
plt.plot(errors, final, label="l1")

final = [
    l2_combiner([sigmoid, softplus, clipped_linear])
    for sigmoid, softplus, clipped_linear in zip(
        sigmoid_errors, softplus_errors, clipped_linear_errors
    )
]
plt.plot(errors, final, label="l2")

final = [
    smoothmax_combiner([sigmoid, softplus, clipped_linear])
    for sigmoid, softplus, clipped_linear in zip(
        sigmoid_errors, softplus_errors, clipped_linear_errors
    )
]
plt.plot(errors, final, label="smoothmax")

plt.legend()
plt.show()
```
